## Nikola Tesla

**Nikola Tesla** je najpoznatiji po svom doprinosu u projektovanju modernog sisitema napajanja *naizmeničnom struje*.

Njegovi najznačajniji pronalasci su:
- polifazni sistem
- obrnuto magnetsko polje
- asinhroni motor
- sinhroni motor
- teslin transformator
